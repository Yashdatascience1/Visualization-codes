```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('dublin.csv')
```

```python
df.head()
```

```python
df = df[['Name','Latitude','Longitude']]
```

```python
map = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], zoom_start=14, control_scale=True)
```

```python
map
```

```python
df.head(10)
```

```python
for i,j in df.iterrows():
    folium.Marker(location=[j["Latitude"],j["Longitude"]],popup=j["Name"]).add_to(map)
```

#### Scatterplots between response variable and predictor variables

```python
def response_scatter(df,response):
    fig,axes=plt.subplots(nrows=len(df.drop(response,axis=1).columns),figsize=(10,30))
    m=0
    for i in df.drop(response,axis=1).columns:
        sns.scatterplot(x=df[i],y=df[response],ax=axes[m])
        axes[m].set_title("Scatter plot of "+str(i)+" with "+str(response))
        m+=1
    plt.tight_layout()
```

#### Distribution and qqplot of continuous variables

```python
def dist_qq_and_box_plot(df,cont):
    import statsmodels.api as SM1
    fig,axes=plt.subplots(nrows=len(df[cont].columns),ncols=3,figsize=(10,15))
    m=0
    for i in df[cont].columns:
        n=0
        sns.distplot(df[i],ax=axes[m][n])
        axes[m][n].set_title("Distribution plot of "+str(i))
        n+=1
        SM1.qqplot(df[i],ax=axes[m][n],line="s")
        axes[m][n].set_title("QQ plot of "+str(i))
        n+=1
        sns.boxplot(df[i],ax=axes[m][n])
        axes[m][n].set_title("Box plot of "+str(i))
        m+=1
    plt.tight_layout()
```

### Heatmap of correlation matrix

```python
def corr_heatmap(data):
    cm=data.corr()
    plt.figure(figsize=(10,30))
    sns.heatmap(cm,cbar=True,annot=True,fmt=".2f",annot_kws={"Size":10},lw=0.5,mask=np.triu(cm,+1))
```

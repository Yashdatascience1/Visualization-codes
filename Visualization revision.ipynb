```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y) #plotting the graph
plt.title('Our First Graph!',fontdict={'fontname':'Calibri','fontsize':20})
plt.xticks([0,1,2,3,4]) #labeling the ticks
plt.yticks([0,2,4,4.5,6.4,7,8,10])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show() #written to remove the line which shows the memory location of this function
```
